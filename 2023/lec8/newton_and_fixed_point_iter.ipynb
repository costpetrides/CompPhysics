```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
def f(x): return x**2 - 11*x + 25

def df(x): return 2*x - 11    
```

```python
x = 5

N = 10
for i in range(N):
    x = x - f(x)/df(x)
    print('x = ', x, ' f(x) = ', f(x))
    
```

```python
def g(x) : return 5 + x**0.5
```

```python
x = 1
for i in range(100):
    x = g(x)
    ε = abs(g(x) - x)
    print('x = %2.3f , f(x) = %2.6f'% (x, f(x)))

    if ε <10**-8: break
        
    
```

```python
xs = np.linspace(2,10, 100)
plt.plot(xs, f(xs))
```

```python
plt.plot(xs, g(xs))
plt.plot(xs, xs)
```
