```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def U(): return np.random.rand() # generates randoms in 0, 1
```

```python
def dice():
    u = U()
    res = 0
    if 0    <= u < 1/6.: res = 1 
    if 1/6. <= u < 2/6.: res = 2
    if 2/6. <= u < 3/6.: res = 3  
    if 3/6. <= u < 4/6.: res = 4
    if 4/6. <= u < 5/6.: res = 5
    if 5/6. <= u < 6/6.: res = 6
    return res
       
```

```python
dices = [dice() for i in range(100000)] # list comphrension
# equivalent
# dices = []
# for i in range(100000):
#     dices += [dice()]
    
plt.hist(dices, bins = range(0,10), ec = 'black')
```

```python
def game():
    pos = 1
    n = 0
    for i in range(50000):
        #pos = pos + dice()
        pos += dice()
        if pos == 3: pos = 24
        if pos == 21: pos = 7
    
        #print(pos)
        n += 1
        if pos >= 25: break

    #print('n = ', n)
    return n
```

```python
games = [game() for i in range(50000)]
```

```python
plt.hist(games, bins = range(0,35), ec='black')
```

```python
games = np.array(games)
print('μ = ', games.mean(), 'σ = ', games.std())
```

```python
# to access frequencis and bin ranges
frequencies, bins, patches = plt.hist(games, bins = range(0,62), ec='black')
for n in range(len(frequencies)):
    print('p(%d) = %d / %d = %2.5f'%(n, frequencies[n], sum(frequencies), frequencies[n]/sum(frequencies)))
```
