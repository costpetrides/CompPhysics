```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def U(): return np.random.rand()
```

```python
Us = [U() for i in range(100000)]
# Us = [0.2643939660088278,
#  0.8765630886473887,
#  0.6791970418298072,
#  0.9697391356071046,
#  0.0713704563739136,
#  0.9829084479237447,
#  0.2232106789006828,
#  0.553669388688878,
#  0.8908404294769777,
```

```python
def exp(x): return np.exp(x)
def ln(x) : return np.log(x)

def invF(x): 
    a, b = 1, 4
    return ln(x*(exp(b) - exp(a)) + exp(a))
```

```python
# Xs = []
# for U in Us:
#     X = invF(U)
#     Xs += [X] # append X in list of Xs

# above code is equivalent with the one below
Us = np.array(Us)
Xs = invF(Us)    

# plot histogram
plt.hist(Xs, bins = 100, ec = 'black')    
```

```python
a, b = 1, 4

def f(x): 
    c = (exp(b) - exp(a))**-1
    return c*exp(x)

xs = []
fmax = f(b)

for i in range(100000):
    x = U()*(b - a) + a
    y = U()*fmax
    #print(x, y, f(x))
    if y < f(x):
        #print(x, y)
        xs += [x]
        
plt.hist(xs, bins = 100, ec = 'black')
```
