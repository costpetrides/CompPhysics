```python
import numpy as np 
import matplotlib.pyplot as plt
U   = lambda : np.random.rand()
```

```python
M = 10000 # walkers, physicists
N = 1000 # number of steps

#xs = [0 for _ in range(M)]
xs = []

def step():
    res = -1
    if U()>0.5: res = +1
    return res

for m in range(M):
    x = 0
    for _ in range(N):
        x = x + step()
        
    xs += [x]
```

```python
_ = plt.hist(xs, bins = np.linspace(-150, 150), ec = 'black')
```

```python
M = 1000 # walkers, physicists
N = 1000 # number of steps

#xs = [0 for _ in range(M)]
xs = []

johnny = [0]

def step():
    res = -1
    if U()>0.5: res = +1
    return res

for m in range(M):
    x = 0
    for _ in range(N):
        x = x + step()
        if (m ==1 ): # if johnny
            johnny += [step() + johnny[-1]] 
        
    xs += [x]
```
