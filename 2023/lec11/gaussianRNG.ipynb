```python
import numpy as np 
import matplotlib.pyplot as plt
U   = lambda : np.random.rand()
ln  = lambda x: np.log(x)
sin = lambda x: np.sin(x)
cos = lambda x: np.cos(x)
π   = np.arccos(-1)
```

```python
rs = []
θs = []
xs = [] 
ys = []

σ = 1
for _ in range(1000000):
    r = (-2*(σ**2)*ln(U()))**0.5
    θ = 2*π*U()
    x = r*cos(θ)
    y = r*sin(θ)
    
    rs += [r]
    θs += [θ]
    xs += [x]  
    ys += [y]
```

```python
_ = plt.hist(rs, bins = 100, ec = 'black')
```

```python
_ = plt.hist(θs, bins = 100, ec = 'black')
```

```python
_ = plt.hist(xs, bins = 100, ec = 'black')
```

```python
_ = plt.hist(ys, bins = 100, ec = 'black')
```

```python
_ = plt.hist2d(xs, ys, bins = 100)
```

```python
_ = plt.hexbin(xs, ys, bins = 100)
```
