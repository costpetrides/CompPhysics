```python
import numpy as np
import matplotlib.pyplot as plt 

U = lambda : np.random.rand() # def U(): return np.random.rand()
```

```python
jMax = 100000

S1  = []
S2  = []
S20 = []

for _ in range(jMax):
    X1  = sum([U() for _ in range(1)])/1
    X2  = sum([U() for _ in range(2)])/2
    X20 = sum([U() for _ in range(20)])/20
    
    S1  += [X1]
    S2  += [X2]
    S20 += [X20]
```

```python
_ = plt.hist(S1 , bins = np.linspace(0, 1), ec = 'black')
_ = plt.hist(S2 , bins = np.linspace(0, 1), ec = 'black')
_ = plt.hist(S20, bins = np.linspace(0, 1), ec = 'black')
```
