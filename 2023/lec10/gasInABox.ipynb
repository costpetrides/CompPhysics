```python
import numpy as np
import matplotlib.pyplot as plt 

U = lambda : np.random.rand() # def U(): return np.random.rand()
```

```python
N1 = 10**4
N2 = 0

p1 = 0.05
p2 = 0.01

N1t = []
N2t = []

N1t += [N1]
N2t += [N2]

for _ in range(100):
    dN1 = 0
    dN2 = 0
    for _ in range(N1):
        if (U() < p1): dN1 += 1

    for _ in range(N2):
        if (U() < p2): dN2 += 1
            
    N1 = N1 - dN1 + dN2
    N2 = N2 - dN2 + dN1
    
    N1t += [N1]
    N2t += [N2]
    
print(N1, N2)
_ = plt.plot(N1t)        
_ = plt.plot(N2t)        
```
